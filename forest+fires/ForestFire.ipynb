```python

import pandas as pd
import matplotlib.pyplot as plt

#  Load the dataset
df = pd.read_csv("forestfires.csv")
print(" Dataset loaded successfully.\n")
```

```python
# Show first few rows
print(" First 5 records:")
print(df.head(), "\n")
```

```python
#  Basic info
print(" Dataset Info:")
print(df.info(), "\n")
```

```python
#  Check for missing values
print(" Missing values per column:")
print(df.isnull().sum(), "\n")
```

```python
#  Statistical summary
print(" Summary Statistics:")
print(df.describe(), "\n")
```

```python
#  Plot number of fires by month
if "month" in df.columns:
    month_counts = df["month"].value_counts().sort_index()
    plt.figure(figsize=(10, 5))
    plt.bar(month_counts.index, month_counts.values, color='forestgreen')
    plt.title("Number of Forest Fires per Month")
    plt.xlabel("Month")
    plt.ylabel("Number of Fires")
    plt.grid(True)
    plt.show()
```

```python

#  Area burned distribution
if "area" in df.columns:
    plt.figure(figsize=(10, 5))
    plt.hist(df["area"], bins=30, color='darkred', edgecolor='black')
    plt.title("Distribution of Burned Area")
    plt.xlabel("Area (hectares)")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()
```

```python
#  Scatter plot between temperature and area burned
if "temp" in df.columns and "area" in df.columns:
    plt.figure(figsize=(8, 5))
    plt.scatter(df["temp"], df["area"], alpha=0.6, color='orange')
    plt.title("Temperature vs Burned Area")
    plt.xlabel("Temperature (C)")
    plt.ylabel("Burned Area (hectares)")
    plt.grid(True)
    plt.show()
```

```python
#  Scatter plot between wind and area burned
if "wind" in df.columns and "area" in df.columns:
    plt.figure(figsize=(8, 5))
    plt.scatter(df["wind"], df["area"], alpha=0.6, color='blue')
    plt.title("Wind Speed vs Burned Area")
    plt.xlabel("Wind Speed (km/h)")
    plt.ylabel("Burned Area (hectares)")
    plt.grid(True)
    plt.show()
```

```python
#  Insights summary
print(" Analysis complete!")
if "month" in df.columns:
    print(" Peak fire month:", month_counts.idxmax())
if "temp" in df.columns:
    print(" Average temperature:", round(df["temp"].mean(), 2), "C")
if "area" in df.columns:
    print(" Average burned area:", round(df["area"].mean(), 2), "hectares")
```
